```python
# a longwinded way to create a list of the neighbourhood codes

list_nbh = str(force.neighbourhoods).replace('<Neighbourhood>', '')
list_nbh2 = list_nbh.replace(',', '')
apinbh = list_nbh2.split()
apinbh.remove('[')
apinbh.remove('E05013584]')
apinbh.append('E05013584')
```

```python
# a simple function which retrieves the Neighbourhood name of each LSOA code 

def full_met_nbh(LSOAcode):
    if LSOAcode in apinbh:
        neighbourhood = (Neighbourhood(api, force=force, id=LSOAcode))
        nname = neighbourhood.name
        return nname
    else:
        return 'n/a'
```

The function aimed to convert the LSOA codes into the neighbourhoods. While the code ran, another issue became clear. After extensive research, including reading a Freedom of Information Act, it was found that the Met police use 'neighbourhoods' as one geographical division, and 'LSOA codes' as another - there is no crossover/link between these LSOA codes and the 'neighbourhoods'.
