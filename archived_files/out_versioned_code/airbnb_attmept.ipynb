## June 2022 and March 2023 Airbnb listings files were cleansed separately:

```python
import pandas as pd
```

```python
raw_june_airbnb_listings = pd.read_csv('/content/gdrive/MyDrive/Group 7 Files/data/csv/airbnb_datasets/airbnb_files_raw_uncleansed/detailed_listings_scraped_June22.csv')
column_names = raw_june_airbnb_listings.columns.values.tolist()
column_names
```

```python
jun_airbnb = raw_june_airbnb_listings.copy()
jun_airbnb = raw_june_airbnb_listings[['id','last_scraped','neighbourhood_cleansed', 'latitude', 'longitude', 'property_type', 'room_type', 'price']]
jun_airbnb.rename(columns = {'neighbourhood_cleansed':'borough'})
```

```python
len(jun_airbnb)-jun_airbnb.count() #this tells us there are no null values in our data
```

```python
jun_airbnb.to_csv('cleaned_detailed_listings_scraped_June22.csv')
```

```python
from google.colab import files 
files.download('cleaned_detailed_listings_scraped_June22.csv')
```

```python
jun_airbnb_neighbourhood_count = jun_airbnb.neighbourhood_cleansed.value_counts().reset_index().rename(
           columns={'index': 'borough', 'neighbourhood_cleansed': 'count'})
jun_airbnb_neighbourhood_count
```

```python
jun_airbnb_neighbourhood_count.to_csv('cleaned_airbnb_neighbourhood_count_June22.csv')
```
