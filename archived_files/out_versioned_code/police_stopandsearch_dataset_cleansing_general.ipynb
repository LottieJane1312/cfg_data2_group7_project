```python
stop_and_search_df_copy = stop_and_search_df.copy()
stop_and_search_df_copy = stop_and_search_df[['Type', 'Longitude', 'Latitude','Object of search', 'Outcome']]
stop_and_search_df_copy
```

```python
missing_values = stop_and_search_df_copy.isnull().sum()

for column, count in missing_values.items():
    formatted_count = "{:,}".format(count)
    print(f"Column '{column}' has {formatted_count} missing values")
    
print('Therefore, we want to remove the entries with no locations as this will not be useful in our analysis')
```

```python
stop_and_search_df_cleaned = stop_and_search_df_copy.dropna(axis=0, subset=['Longitude'])
```
