## Lottie modified Trupti's code to fit in & remove the remaining LSOA Names that aren't in London
### Trupti's original notebook can be found in the archived_files named 'Crime_march_exploratory'

```python
# Store the airbnb neighbourhoods to a new variable 
lsoa_names_df = airbnb_neighbourhoods

# Create a new column in police_street_combined_data_copy called "matched_neighbourhood" and initialize it with "not_found"
police_street_combined_data_copy['matched_neighbourhood'] = 'not_found'

# Iterate over each row in police_street_combined_data_copy and compare the "LSOA name" with the modified "neighbourhood" column
for index, row in police_street_combined_data_copy.iterrows():
    lsoa_name = row['LSOA name']
    matching_neighbourhood = lsoa_names_df[lsoa_names_df['neighbourhood'].str.contains(lsoa_name, regex=False)]
    if not matching_neighbourhood.empty:
        police_street_combined_data_copy.at[index, 'matched_neighbourhood'] = matching_neighbourhood['neighbourhood'].values[0]
```

```python
# Let's quickly check that the remaning not_found columns are not in the London Boroughs before we drop them from the dataset - CONFIRMED 

filtered_data = police_street_combined_data_copy[police_street_combined_data_copy['matched_neighbourhood'] == 'not_found']

filtered_data
```

```python
# Drop the rows that are 'not_found' and therefore NOT a London Borough for our analysis - we will include an explanation of our findings in our report 
# as to why there are locations not inside London inside the London dataset published by the Police 

police_street_combined_data_copy = police_street_combined_data_copy[police_street_combined_data_copy['matched_neighbourhood'] != 'not_found']
```
