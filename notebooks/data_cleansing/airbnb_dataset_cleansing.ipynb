```python
# Compare column names as there are not equal columns above 

if airbnb_june22.columns.equals(airbnb_march23.columns):
    print("Column names match.")
else:
    print("Column names do not match.")
```

```python
# COLUMNS DO NOT MATCH - join the datasets together into one dataframe ON COLUMN NAME and reset the index 

dfs = [airbnb_june22, airbnb_march23]
airbnb_combined_data = pd.concat(dfs, axis=0, ignore_index=True)
```

```python
# check that there are 33 London Boroughs 

num_unique_boroughs = airbnb_combined_data['neighbourhood_cleansed'].nunique()
print(f"The number of unique values in the 'neighbourhood_cleansed' column is: {num_unique_boroughs}")
```

```python
# copy the dataset from here so we don't mess up what we've done above 

airbnb_data_copy = airbnb_combined_data.copy()
```

```python
# choose the columns that we want to keep in the dataset - after a group meeting where we explored the data collectively it was decided to keep the following columns in the dataset

airbnb_data_copy = airbnb_data_copy[['id','last_scraped', 'host_id','host_since', 'host_location', 'host_total_listings_count', 'neighbourhood_cleansed', 'latitude', 'longitude', 'property_type', 'room_type', 'accommodates', 'bathrooms_text', 'bedrooms', 'beds', 'price', 'number_of_reviews', 'review_scores_rating', 'minimum_nights', 'maximum_nights', 'has_availability', 'calculated_host_listings_count']]
```

```python
# rename the neighbourhood cleansed column to borough

airbnb_data_copy.rename(columns={'neighbourhood_cleansed': 'borough'}, inplace=True)
```
